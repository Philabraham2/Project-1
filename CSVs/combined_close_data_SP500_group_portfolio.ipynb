```python
# Importing dependencies and libraries
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
SP500_data = Path("SP500_close_data.csv")

portfolio_data = Path("raw_portfolio.csv")

SP500_df = pd.read_csv(SP500_data, index_col=0, parse_dates=True, infer_datetime_format=True)
portfolio_df = pd.read_csv(portfolio_data, index_col=0, parse_dates=True, infer_datetime_format=True)


SP500_df
#portfolio_df
```

```python
#concat the df's

joint_portfolio = pd.concat([SP500_df, portfolio_df], axis='columns', join='outer')

joint_portfolio.dropna()
```

```python
column_names = joint_portfolio.columns.values
column_names[0] = 'SP500'
joint_portfolio.columns = column_names


joint_portfolio.dropna(inplace=True)

joint_portfolio
```

```python
joint_portfolio.to_csv(r'combined_close_data_SP500_group_portfolio', index=True, header=True)
```
